```python
import pandas as pd
import numpy as np
Apr_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\april22.csv")
May_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\May2022.csv")
June_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\June22.csv")
July_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\July22.csv")
August_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\August22.csv")
September_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\September22.csv")
October_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\October22.csv")
November_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\November22.csv")
December_22=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\December22.csv")
January_23=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\January2023.csv")
February_23=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\February23.csv")
March_23=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\March23.csv")
April_23=pd.read_csv(r"C:\Users\PRAVEEN CHANDRA\Desktop\April23.csv")
year=pd.concat([Apr_22,May_22,June_22,July_22,August_22,September_22,October_22,November_22,December_22,January_23,February_23,March_23,April_23],axis=0)
year=year.drop(['start_station_name','start_station_id','end_station_name','end_station_id','start_lat','start_lng','end_lat','end_lng','start_day','ride_length'],axis=1)
year.describe()

```

```python
#locating and checking duplicates
year.loc[year.duplicated('ride_id'),:]
```

```python
YEAR=year.drop_duplicates(subset=['ride_id'])
YEAR
```

```python
year.info()
```

```python
YEAR.loc[:,'started_at']=pd.to_datetime(YEAR.loc[:,'started_at'],format='%d-%m-%Y %H:%M:%S')
YEAR.loc[:,'ended_at']=pd.to_datetime(YEAR.loc[:,'ended_at'],format='%d-%m-%Y %H:%M:%S')
YEAR.loc[:,'ride_duration']=YEAR.loc[:,'ended_at']-YEAR.loc[:,'started_at']
pd.set_option('display.max_rows',None)
YEAR
```

```python
YEAR['start_day'] =YEAR['started_at'].dt.day_name()
YEAR['start_month'] =YEAR['started_at'].dt.month_name()
YEAR=YEAR.sort_values(['started_at'])
```

```python
#average ride length of users by membership
YEAR.groupby('member_casual')['ride_duration'].mean()
```

```python
#number of rides of users by day of week
YEAR.groupby('start_day')['ride_id'].count()
```

```python
#average ride length by day of week for membership
YEAR.groupby(['start_day','member_casual'])['ride_duration'].mean()
```
